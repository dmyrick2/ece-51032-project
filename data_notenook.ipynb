```python
# ELECTRICITY COST DATA
import pandas as pd
# data day ahead: http://www.energyonline.com/Data/GenericData.aspx?DataId=23&ERCOT___Day-Ahead_Price
# data real time: http://www.energyonline.com/Data/GenericData.aspx?DataId=23&ERCOT___Day-Ahead_Price

hourly_dayahead = pd.read_csv("2019_ercot_day_ahead_price.csv")
hourly_realtime = pd.read_csv("2019_ercot_real_time_price.csv")
hourly_dayahead_realtime = pd.merge(hourly_dayahead, hourly_realtime, left_on=["zone", "Date"], right_on=["zone","Date"], how='inner', suffixes=("_dayahead", "_realtime"))
hourly_dayahead_realtime["Date"] = pd.to_datetime(hourly_dayahead_realtime["Date"])
solar_data = pd.read_csv("solar_profile_29_-95.csv", header=3)
solar_data['local_time'] = pd.to_datetime(solar_data['local_time'])
hourly_dayahead_realtime_solar = pd.merge(hourly_dayahead_realtime, solar_data, left_on="Date", right_on="local_time", how="left")
```
